```python
# pandas -> analise de dados
# getting started
# quickstart
import pandas as pd

df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
})


display(df)
print(df["Age"])
# pd.DataFrame
# pd.DataFrame -> recebe um dicionário como parâmetro e cria a tabela a partir do dicionário
```

```python
ages = pd.Series([22, 35, 58], name="Age")

print(df["Age"].max())
print(ages.max())
```

```python
# datas no pandas
import pandas as pd

tabela = pd.read_excel("Contratos.xlsx")
tabela['Data Pagamento'] = pd.to_datetime(tabela['Data Pagamento'], format='%d/%m/%Y')
tabela['Data Pagamento 2'] = pd.to_datetime(tabela['Data Pagamento 2'], format='%m-%d-%y')
print(tabela.dtypes)


display(tabela)
```

```python
# funcao merge do pandas

emails = pd.read_excel("Emails.xlsx")
tabela = tabela.merge(emails, left_on="Cliente", right_on="Nome")
display(tabela)
```
